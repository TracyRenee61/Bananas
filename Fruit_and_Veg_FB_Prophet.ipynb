Import librariesd

```
import numpy as np
import pandas as pd
```

Read csv file

```
data = pd.read_csv('https://raw.githubusercontent.com/TracyRenee61/Bananas/main/fruitvegprices-11jan21.csv', encoding= 'unicode_escape')
data
```

```
data.info()
```

Check for null values

```
data.isnull().sum()
```

Sort dataframe by date

```
data = data.sort_values(by='date').reset_index()
data
```

Create group

```
data['Group'] = data['item'] + "-" + data['variety']
data
```

Drop columns

```
data = data.drop(['item', 'variety'], axis=1)
data
```

Count group

```
groups = data['Group'].to_list()

# using naive method to remove duplicated from list  
list_group = [] 
for i in groups: 
    if i not in list_group: 
        list_group.append(i) 
sorted(list_group)
```

```
print(len(list_group))
```

Dates

```
from datetime import date 
from datetime import timedelta
```

```
max_date = data.date.max()
max_date = pd.Timestamp(max_date)
max_date
```

Rename columns

```
data = data.drop(['unit', 'index'], axis=1)

df = data.rename(columns={'date': 'ds', 'group': 'group', 'price':'y'})
df
```

FB Prophet predictions on each group

```
from fbprophet import Prophet

final = pd.DataFrame()

grouped = df.groupby('Group')
for g in grouped.groups:
    group = grouped.get_group(g)
    m = Prophet()
    m.fit(group)
    future = m.make_future_dataframe(periods=52, freq='w')
    forecast = m.predict(future)
    #I added a column with group
    forecast['group'] = g
    #I used concat instead of merge
    final = pd.concat([final, forecast], ignore_index=True)
final
```

Get predictions

```
prediction = pd.DataFrame({'date': final.ds, 'prediction': final.yhat, 'group':final.group })
prediction
```

Get future predictions

```
future_prediction = prediction[(prediction['date'] > max_date)]
future_prediction
```

Count groups from future predictions

```
groups = future_prediction['group'].to_list()

# using naive method to remove duplicated from list  
list_group = [] 
for i in groups: 
    if i not in list_group: 
        list_group.append(i) 
sorted(list_group)
```

```
print(len(list_group))
```

Select group

```
origin = 'turnip-turnip'

single_prediction = future_prediction[(future_prediction['group'] == origin)]
single_prediction
```
